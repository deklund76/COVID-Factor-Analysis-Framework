This database cleaner is for:

United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv

Steps to clean the file:

(**Name of relevant dataframe:  df, df_cases, df_deaths**)

1.  Chronological.  Change column "submission_date" dtype to datetime format. Place column "submission" in increasing order AND sort by "state" also.


2.  Column deletion.  Delete column "created_at".


3.  Retaining rows.  Keep row if value of either "consent_cases" or "consent_deaths" is Agree or Not Agree.


4.  Fill in missing values.  Replace all null values with the integer 0 for columns "tot_cases", "conf_cases", "prob_cases", "new_case", "pnew_case", "tot_death", "conf_death", "prob_death", "new_death", "pnew_death".


5.  Change datatype.  For columns "tot_cases", "conf_cases", "prob_cases", "new_case", "pnew_case", "tot_death", "conf_death", "prob_death", "new_death", "pnew_death", change dtype to integer.


6.  Apply OneHotEncoder.  At this poin columns "consent_cases" or "consent_deaths" will have values Agree, Not Agree, or N/A.  Apply OneHotEncoder to THREE columns:  "consent_cases", "consent_deaths", and "state."


7.  Assume that only the year is needed.  Extract the year and place it in a column "Year_submitted."  Delete the column "submission_date."

8.  Make a new dataframe for cases only.

9.  Make a new dataframe for deaths only.

10.  Perform a statistical analysis for columns "tot_cases" and "tot_death."


    (**Name of relevant dataframes:  df2, df2_cases_stats, df2_deaths_stats**)

    a.  Read "United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv" in again and give it a differenct name.
    
    b.  Perform a describe() on column "tot_cases" over the year 2020 only for all states.  The results apply to the           population of states.
    
    c.  Perform a describe() on column "tot_death" over the year 2020 only for all states.  The results apply to the           population of states.
    
    d.  Select a statistical measure to describe an assessment column for states relative to the population.  For             *starters* use the population mean.  For example, if the number of cases for a state is above or equal to the         population mean assign a 1 to the assessment column, and if less than the population mean assign a 0 for               the state to the assessment column.  An assessment column for cases will be assigned to the cases dataframe           and an assessment column for deaths will be assigned to the deaths dataframe.  
    
    e.  Apply the 2020 population mean assessment criterion to both the years 2020 and 2021. In this way the states           will be assessed against the previous year (2020) for testing.  *This means that training will done be on 2020         data and that the testing will be done on 2021 data (but using 2020 assessment criteria for assessing 2021             relative to 2020).*

11.  Add a columnn ">= 2020 mean" to df_cases.

12.  Populate ">= 2020 mean" with 1 or 0 as described above.

13.  Add a columnn ">= 2020 mean" to df_deaths.

14.  Populate ">= 2020 mean" with 1 or 0 as described above.

15.  Save df_cases as csv file.

16.  Save df_deaths as csv file.


Summary of dataframes used for this cleaning script:

**df**:  this is the *first* time "United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv" is read

**df_cases**:  this is made from df, and a new column ">= 2020 mean" is added to it based on statistical analysis from                df2_cases_stats

**df_deaths**: this is made from df, and a new column ">= 2020 mean" is added to it based on statistical analysis from                df2_cases_stats

**df2**:   this is the *second* time "United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv" is read

**df2_cases_stats**: this is from df2 and holds only columns needed for statistical analysis

**df2_deaths_stats**:  this is from df2 and holds only columns needed for statistical analysis


**This file is only for cleaning data to be used for input to a machine learning model.**  Machine learing models will have separate file(s) but will read in dataframes "df_cases" and "df_deaths."

```python
#Import dependencies

import pandas as pd
```

```python
# read the file

file_path = "Resources/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv"
df = pd.read_csv(file_path)
df.head()
```

**Step 1**

```python
# Change column "submission" dtype to datetime format.

df["submission_date"] = pd.to_datetime(df["submission_date"])
df.head()
```

```python
# Place column "submission" in increasing order AND sort by "state" also.

df = df.sort_values(by = ["submission_date", "state"])
df.head(200)
```

**Step 2**

```python
# Delete column "created_at".

df.drop(columns=["created_at"], inplace = True)
df.head()
```

**Step 3**

```python
# Keep row if value of either "consent_cases" or "consent_deaths" is Agree or Not Agree.
df = df.loc[(df["consent_cases"]=="Agree")|(df["consent_deaths"]=="Agree") ]
df.head(200)
```

**Step 4**

```python
# Fill in missing values.  Replace all null values with the integer 0 for columns "tot_cases", "conf_cases",
#"prob_cases", "new_case", "pnew_case", "tot_death", "conf_death", "prob_death", "new_death", "pnew_death".

zero_list = ["tot_cases", "conf_cases","prob_cases", "new_case","pnew_case", "tot_death","conf_death", "prob_death","new_death", "pnew_death"]

for x in zero_list:
    df[x] = df[x].fillna(0)

df.head(200)
```

**Step 5**

```python
# For columns "tot_cases", "conf_cases", "prob_cases", "new_case", "pnew_case", "tot_death", "conf_death",
# "prob_death", "new_death", "pnew_death", change dtype to integer.


col_headers = ["tot_cases", "conf_cases", "prob_cases", "new_case", "pnew_case", "tot_death", "conf_death",
"prob_death", "new_death", "pnew_death"]

for col in col_headers:
    df[col]=df[col].astype("int64")

df.head(200)
```

```python
# import dependencies

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
obj_list = df.dtypes[df.dtypes == "object"].index.to_list()
obj_list
```

```python
# Apply OneHotEncoder to THREE columns:  "consent_cases", "consent_deaths", and "state."

enc = OneHotEncoder(sparse = False)
encoded_df = pd.DataFrame(enc.fit_transform(df[obj_list]))
encoded_df.columns = enc.get_feature_names(obj_list)
encoded_df.head(200)
```

```python
df = df.merge(encoded_df, left_index = True, right_index = True)
df = df.drop(obj_list, 1)
df.head(200)
```

**Step 10**

```python
#  perform a statistical analysis for columns "tot_cases" and "tot_death."

# Read "United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv" in again and give it a differenct name.

file_path = "Resources/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv"
df2 = pd.read_csv(file_path)
df2.head()
```

```python
df2.shape
```

```python
df2_headers = df2.columns.to_list()
df2_headers
```

```python
# Perform a describe() on column "tot_cases" over the year 2020 only for all states.  The results apply to the 
# population of states.


df2_cases_stats = df2.copy()
df2_cases_stats.drop(columns = ['conf_cases','prob_cases','new_case','pnew_case','tot_death','conf_death',
'prob_death','new_death','pnew_death','created_at','consent_cases','consent_deaths'], inplace = True)
df2_cases_stats["submission_date"] = pd.to_datetime(df2_cases_stats["submission_date"])
df2_cases_stats["tot_cases"] = df2_cases_stats["tot_cases"].fillna(0)
df2_cases_stats["tot_cases"] = df2_cases_stats["tot_cases"].astype("int64")
df2_cases_stats = df2_cases_stats.sort_values(by=["submission_date","state"] )
df2_cases_stats["Year_submitted"] = df2_cases_stats["submission_date"].dt.year
df2_cases_stats.drop(columns = ["submission_date"], inplace = True)
df2_cases_stats = df2_cases_stats.reindex(columns = ["Year_submitted", "state", "tot_cases"])

# find only 2020 data for cases

df2_cases_stats = df2_cases_stats.loc[df2_cases_stats["Year_submitted"]==2020]
df2_cases_mean = df2_cases_stats["tot_cases"].mean()
df2_cases_mean
```

```python
# Perform a describe() on column "tot_death" over the year 2020 only for all states.  The results apply to the 
# population of states.


df2_deaths_stats = df2.copy()
df2_deaths_stats.drop(columns = ['tot_cases', 'conf_cases','prob_cases','new_case','pnew_case','conf_death',
'prob_death','new_death','pnew_death','created_at','consent_cases','consent_deaths'], inplace = True)
df2_deaths_stats["submission_date"] = pd.to_datetime(df2_deaths_stats["submission_date"])
df2_deaths_stats["tot_death"] = df2_deaths_stats["tot_death"].fillna(0)
df2_deaths_stats["tot_death"] = df2_deaths_stats["tot_death"].astype("int64")
df2_deaths_stats = df2_deaths_stats.sort_values(by=["submission_date","state"] )
df2_deaths_stats["Year_submitted"] = df2_deaths_stats["submission_date"].dt.year
df2_deaths_stats.drop(columns = ["submission_date"], inplace = True)
df2_deaths_stats = df2_deaths_stats.reindex(columns = ["Year_submitted", "state", "tot_death"])

# find only 2020 data for deaths

df2_deaths_stats = df2_deaths_stats.loc[df2_deaths_stats["Year_submitted"]==2020]
df2_deaths_mean = df2_deaths_stats["tot_death"].mean()
df2_deaths_mean
```

**Steps 10d, 10e, 11, 12, 13, 14**

```python
# Select a statistical measure to describe an assessment column for states relative to the population. 
# For starters use the population mean.  For example, if the number of cases for a state is above or equal
# to the population mean assign a 1 to the assessment column, and if less than the population mean
# assign a 0 for the state to the assessment column.  An assessment column for cases will be assigned to 
# the cases dataframe and an assessment column for deaths will be assigned to the deaths dataframe.

# Apply the 2020 population mean assessment criterion to both the years 2020 and 2021. In this way the states
# will be assessed against the previous year (2020) for testing.  This means that training will done be on 2020
# data and that the testing will be done on 2021 data (but using 2020 assessment criteria for assessing 2021
# relative to 2020).

# Add a columnn ">= 2020 mean" to df_cases.

# Populate ">= 2020 mean" with 1 or 0 as described above.

# Add a columnn ">= 2020 mean" to df_deaths.

# Populate ">= 2020 mean" with 1 or 0 as described above.


# cases

tot_cases_list = df2["tot_cases"].to_list()

for x in tot_cases_list:
    if x >= int(df2_cases_mean):
        df_cases["2020_mean_cases"] = 1
    else:
        df_cases["2020_mean_cases"] = 0


# deaths

tot_deaths_list = df2["tot_death"].to_list()

for x in tot_deaths_list:
    if x >= int(df2_deaths_mean):
        df_deaths["2020_mean_deaths"] = 1
    else:
        df_deaths["2020_mean_deaths"] = 0

#df_cases.head()
#df_deaths.head()
```
